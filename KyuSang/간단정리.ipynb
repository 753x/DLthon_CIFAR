Dropout을 추가하지 않고 단순하게 구했을 때의 Test Accuracy랑 Dropout을 추가했을 때 Test Accuracy값을 구해봤다.
그 결과 Dropout값이 작을때(Dropout = 0.2) Test Accuracy이가 가장 컸고 Dropout값이 커질수록 Test Accuracy이 점점 작아져 아무것도 추가하지 않았을때보다 Test Accuracy이 작아졌다. 즉, Dropout을 모델에 추가하면 일반적으로 모델의 복잡성을 줄이고, 과적합을 줄이는데 도움이 될 수 있지만 Dropout 비율을 너무 높게 설정하면 모델이 훈련 중에 너무 많은 뉴런을 무작위로 비활성화하여 정보를 잃을 수 있게되어 모델의 성능을 저하가 될수 있음을 확인할 수 있다.

# 피드백 및 느낀점

1) 드룹아웃을 하더라도 내가 가지고 있는 데이터에 따라서 값을 알맞게 조정해야한다는 것을 알 수 있었다. 즉, 데이터 양이 충분하지 않으면 Dropout이 더 부정적인 영향을 미칠 수 있다.
2) 드룹아웃을 하면서 0.8일 경우에는 확실히 모델 성능이 저하될것을 예상할 수 있었지만 0.5도 마찬가지로 저하될것이라고 예상하지 못했다.
